# Naša vaja: $H \rightarrow \mu\mu$

- Porazdelitev invariantne mase dveh mionov, ki nastaneta pri razpadu Higgsovega bozona.  
- Particle Data Group (PDG): https://pdglive.lbl.gov/DataBlock.action?node=S126SMU.  
- V nalogi bomo obravnavali (poenostavljeno) meritev eksperimenta ATLAS.  
- Ogledali si bomo kinematično porazdelitev dogodkov po rekonstruirani invariantni masi dveh izbranih mionov:  

$$
m_H = m_{\mu\mu} = \sqrt{(E_{\mu^+} + E_{\mu^-})^2 - |\vec{p}_{\mu^+} + \vec{p}_{\mu^-}|^2} 
$$

- Iskanje signala v ogromnih količinah podatkov je velik izziv.
- Uporaba najnaprednejših računskih metod za ločevanje procesov signala od ozadja.

## Kako prebrati histograme?

- Potrebovali bomo:
    - `bin_edges`: robne $x$ vrednosti binov,
    - `bin_centers`: centralne $x$ vrednosti binov,
    - `bin_values`: število dogodkov za posamični bin,
    - `bin_errors`: pravilne napake za posamični bin.
- Napaka na binih za simulirano ozadje in signal je že določena: $\sigma^2_k = \sum_{x_i \in k} w_i^2$.
- Napaka na binih podatkov je Poissonska: $\sigma_k^2 = N_k$.
- Pri delu se namenoma zmoti, preveri učinek Poissonske in konstantne napake na simuliranem ozadju in signalu.

````python
def load_histogram(hist_file):
    with np.load(hist_file, "rb") as data:
        bin_edges = data["bin_edges"]
        bin_centers = data["bin_centers"]
        bin_values = data["bin_values"]
        bin_errors = data["bin_errors"]

    return {"centers": bin_centers, "edges": bin_edges, "values": bin_values, "errors": bin_errors}
````

````python
hists = {}

for hist_label, hist_file in hists_files.items():
    hists[hist_label] = load_histogram(hist_file)
````

## Modeliranje ozadja

- Razmerje signal/ozadje je $\sim$ 0.2\% v ombočju $m_{\mu\mu}\in[120, 130]$ GeV $\Rightarrow$ potrebujemo dober opis ozadja za razločevanje med procesi.
- Glavni proces ozadja je razpad $Z\rightarrow\mu^+\mu^-$ (Drell-Yan), ki prispeva resonančno porazdelitev z vrhom pri $m_Z=91.2$ GeV.
- Tudi veliko drugih procesov ozadja (npr. tvorba in razpad $ZZ, WW \text{ in } WZ$, $t\bar{t}$ nastanek, ...)

- Na voljo imamo več metod za ustrezno določitev simulacije ozadja:
    - uteževanje s polinomi nizke stopnje,
    - teoretično motivirani nastavki za modeliranje oblike porazdelitve,
    - uporaba SVM metod z različnimi jedri in regularizatorji,
    - uporaba metod strojnega učenja (npr. GPR),
    - dekompozicija na ortogonalne polinome.

### Glajenje ozadja s funkcijskimi nastavki

- Glajenja ozadja (smoothing) na intervalu 110 GeV naprej, preveri različne funkcije:
    - polinomi: $f(m)=a_1 + a_2m + a_3m^2 + \ldots$,
    - potenčne funkcije: $f(m)=am^b$,
    - eksponentne funkcije: $f(m)=a_1 e^{-b_1m} + a_2 e^{-b_2m} + \ldots$,
    - dijet funkcija: $f(m)=a (1-m^d)^c m^{b_1 + b_2 \log m + b_3 \log m^2 + \ldots}$,
    - eksponent polinoma: $f(m)=a e^{b_1 m + b_2 m^2 + \ldots}$.
- Za fitanje splošnih funkcij bomo uporabili `scipy.optimize.curve_fit`.
- Za fitanje polinonomov lahko uporabimo tudi `np.polyfit`.

#### Primer: fitanje polinomov različnih stopenj

````python
bin_centers, bin_edges, bin_values, bin_errors = hists["Background"].values()

poly_degs = np.arange(2, 11, 2)
poly_results, chi2_results = [], []

for deg in poly_degs:
    coef = np.polyfit(bin_centers, bin_values, deg, w=1 / bin_errors)
    poly = np.poly1d(coef)
    
    fit_y = poly(bin_centers)
    poly_results.append(fit_y)
    
    chi2 = np.sum((bin_values - fit_y)**2 / bin_errors**2)
    chi2_results.append(chi2)
````

````python
fig, ax = plt.subplots(figsize=(8, 7))

for deg, poly_result, chi2_result in zip(poly_degs, poly_results, chi2_results):
    ax.plot(bin_centers, poly_result, label=rf"poly. deg.: {deg}, $\chi^2$: {chi2_result:.2e}")

bkg_hist = hists["Background"]
plot_simple_histogram(ax, bkg_hist["centers"], bkg_hist["values"], bkg_hist["edges"], lw=2, color="k", label="Background")

ax.set_xlabel(r"$m_{\mu\mu}$ [GeV]", fontsize=16)
ax.set_ylabel("$N$", fontsize=16)

ax.legend()
ax.set_yscale("log")
````

### Teoretično motivirani nastavki

- Uporaba različnih teoretično podkrepljenih nastavkov.
- Kolaboracija CMS uporablja nastavek (modificiran Breit-Wigner):\
$$
 mBW(m_{\mu\mu}| m_Z, \Gamma_Z, a_1, a_2, a_3) = \frac{e^{a_2m_{\mu\mu}+a_3m_{\mu\mu}^2}}{(m_{\mu\mu}-m_Z)^{a_1} + (\Gamma_Z/2)^{a_1}} \>.
$$
- Kolaboracija ATLAS uporabi bolj komplicirano funkcijo, ki je v skripti `atlas_fit_function`.
- ATLAS funkcije podajajo zgolj obliko - ne pozabite dodati prostega parametra ali poljubne funkcije (npr. utež s polinomom) za normalizacijo!

#### Primer: fitanje CMS nastavka

````python
from scipy.optimize import curve_fit
````

````python
def cms_fit_function(m, a_1, a_2, a_3):
    m_Z, g_Z = 91.1876, 2.4952
    return np.exp(a_2 * m + a_3 * m**2) / ((m - m_Z) ** a_1 + (0.5 * g_Z) ** a_1)
````

````python
p0 = np.array([1.0, 1.0, 1.0]) * 1e-3
popt, pcov = curve_fit(cms_fit_function, bin_centers, bin_values, sigma=bin_errors, p0=p0)
````

````python
std = np.sqrt(np.diag(pcov))
fit_values = cms_fit_function(bin_centers, *popt)
````

````python
fig, ax = plt.subplots(figsize=(8, 7))

bkg_hist = hists["Background"]
plot_simple_histogram(ax, bkg_hist["centers"], bkg_hist["values"], bkg_hist["edges"], lw=2, color="k", label="Background")

ax.plot(bin_centers, fit_values, c="g", lw=2, label="CMS fit")

ax.set_xlabel(r"$m_{\mu\mu}$ [GeV]", fontsize=16)
ax.set_ylabel("$N$", fontsize=16)

ax.legend()
ax.set_yscale("log")
````

## Modeliranje signala

- V SM je $H\rightarrow\mu\mu$ signal ozka resonanca širine 4.1 MeV pri $m_H=125.09$ GeV.
- Signal je vsota vseh načinov nastanka Higgsovega bozona ($gg$F, VBF, $VH$, $t\bar{t}H$).
- Za parametrizacijo signala se uporabi Crystal Ball (CB) funkcijo, ki je kombinacija Gaussovskega jedra in debelejših potenčnih repov:
$$
\mathrm{CB}\left(m_{\mu \mu} \mid \alpha_{\mathrm{L}, \mathrm{R}}, n_{\mathrm{L}, \mathrm{R}}, m_{\mathrm{CB}}, \sigma_{\mathrm{CB}}\right) =
\begin{cases}
    \mathrm{e}^{-\left(m_{\mu \mu}-m_{\mathrm{CB}}\right)^{2} / 2 \sigma_{\mathrm{CB}}^{2}}, \quad-\alpha_{\mathrm{L}}<\frac{m_{\mu \mu}-m_{\mathrm{CB}}}{\sigma_{\mathrm{CB}}}<\alpha_{\mathrm{R}}                                                                                                                                                                                                    \\
    \left(\frac{n_{\mathrm{L}}}{\left|\alpha_{\mathrm{L}}\right|}\right)^{n_{\mathrm{L}}} \mathrm{e}^{-\alpha_{\mathrm{L}}^{2} / 2}\left(\frac{n_{\mathrm{L}}}{\left|\alpha_{\mathrm{L}}\right|}-\left|\alpha_{\mathrm{L}}\right|-\frac{m_{\mu \mu}-m_{\mathrm{CB}}}{\sigma_{\mathrm{CB}}}\right)^{-n_{\mathrm{L}}}, \frac{m_{\mu \mu}-m_{\mathrm{CB}}}{\sigma_{\mathrm{CB}}} \leq-\alpha_{\mathrm{L}} \\
    \left(\frac{n_{\mathrm{R}}}{\left|\alpha_{\mathrm{R}}\right|}\right)^{n_{\mathrm{R}}} \mathrm{e}^{-\alpha_{\mathrm{R}}^{2} / 2}\left(\frac{n_{\mathrm{R}}}{\left|\alpha_{\mathrm{R}}\right|}-\left|\alpha_{\mathrm{R}}\right|+\frac{m_{\mu \mu}-m_{\mathrm{CB}}}{\sigma_{\mathrm{CB}}}\right)^{-n_{\mathrm{R}}}, \frac{m_{\mu \mu}-m_{\mathrm{CB}}}{\sigma_{\mathrm{CB}}} \geq \alpha_{\mathrm{R}}
\end{cases}
$$
- Na končen rezultat fitamo:
$$
\alpha \cdot \mathrm{CB}\left(m_{\mu \mu} \mid \alpha_{\mathrm{L}, \mathrm{R}}, n_{\mathrm{L}, \mathrm{R}}, m_{\mathrm{CB}}, \sigma_{\mathrm{CB}}\right) \>,
$$
kjer je $\alpha$ normalizacijski faktor.
- **Končni cilj je, da od podatkov odštejete fit ozadja in dobite signal.**
- Da ocenite vaš rezultat, pofitate CB na signal - primer v `fit_CB`.
- **Poskusite najprej parametrizirati/pofitati simuliran signal iz histogramov.**

### Parametrizacija signala s CB funkcijo

- Da pofitate Crystall Ball funkcijo, jo morate obvezno pomnožiti s prostim (*scale*) parametrom (to je `A` v skripti).
- Zgolj s funkcijo `curve_fit` in 7 prostimi parametri boste težko prišli do dobrega fita.
- Ker so prosti parametr $\alpha_{L,R}$ in $n_{L,R}$ visoko korelirani, je za hitro konvergenco smiselno kombinirati samo minimizacijo s pregledom (*scanning*):
    - spreminjaj $n_{L,R}$ po neki mreži smiselnih vrednosti in poišči $\alpha_{L, R}$ z minimizacijo,
    - iterativno obrni postopek: spreminjaj vrednosti $\alpha_{L,R}$ in minimiziraj $n_{L,R}$.
- **Cilj**: faktor za napoved $\hat{\alpha}$ (*ko bomo fitali na odšteto ozadje od podatkov s parametri dobljenimi na MC signalnemu fitu*) čim bližje vrednosti 1.
- Ne pozabi na napake!

````python
from helpers.fit_CB import CrystalBall
````

````python
bin_centers, bin_edges, bin_values, bin_errors = hists["Signal"].values()

popt, pcov = curve_fit(
    CrystalBall,
    bin_centers,
    bin_values,
    sigma=bin_errors,
    p0=[133.0, 1.5, 1.5, 3.7, 9.6, 124.5, 3.0],
)

std = np.sqrt(np.diag(pcov))
fit_values = CrystalBall(bin_centers, *popt)
````

````python
fig, ax = plt.subplots(figsize=(8, 7))
plot_simple_histogram(ax, bin_centers, bin_values, bin_edges, lw=2, color="k", label="Signal")
ax.plot(bin_centers, fit_values, lw=2, color="g", label="CB fit")

ax.set_xlabel(r"$m_{\mu\mu}$ [GeV]", fontsize=16)
ax.set_ylabel("$N$", fontsize=16)

ax.legend()
ax.set_yscale("log")
````
